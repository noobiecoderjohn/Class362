```python
import pandas as pd
import numpy as np
```

Below, I created a small dataset of the sale of each car brand within 4 different cities.

```python
products = ['Honda', 'Toyota', 'Nissan']
regions = ['New York', 'New Jersey', 'Philadelphia', 'Boston']
sales_data = np.random.randint(10000, 50000, size=(len(products), len(regions)))
print(sales_data)
```

Data tidying

```python
sales_df = pd.DataFrame(sales_data, index=products, columns=regions)
print("Number of cars sold by each city:")
print(sales_df)
```

After tidying up everything, I calculated the total sales of each cars by each company and cities.

```python
sales_df = pd.DataFrame(sales_data, index=products, columns=regions)
sales_df['Total Sales'] = sales_df.sum(axis=1)
print("Number of cars sold by each city:")
print(sales_df)
```
