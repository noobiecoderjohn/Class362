```python
delay1 ={
    'Los Angeles': [62], 
    'Phoenix': [12], 
    'San Diego': [20], 
    'San Francisco': [102], 
    'Seattle': [305] 
}
df = pd.DataFrame(delay1)
sum_values = df.sum(axis=1).values[0]  
mean_sum = sum_values / len(df.columns)

# Print the mean of the sum
print("Mean of the delayed flights of all values for Alaska:", mean_sum)
```
