Load the data in the (yellow highlighted) Data
Folder in the file auto-mpg.data into a
pandas DataFrame.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data"
column_names = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model_year', 'origin', 'car_name']
auto_df = pd.read_csv(url, sep=r'\s+', names=column_names)
```

Some of the horsepower data is marked as missing with a ‘?’. You need to replace this ‘?’ with a pandas-appropriate missing value, then convert the column to numericThe origin column values are 1, 2, and 3. 
Convert these values respectively to ‘USA’,‘Asia’, and ‘Europe’
Create a bar chart that shows the distribution for cylinders.

```python
auto_df['horsepower'] = pd.to_numeric(auto_df['horsepower'], errors='coerce')
auto_df['origin'] = auto_df['origin'].map({1: 'USA', 2: 'Asia', 3: 'Europe'})
auto_df['cylinders'].value_counts().plot(kind='bar')
plt.xlabel('Cylinders')
plt.ylabel('Frequency')
plt.title('Distribution of Cylinders')
plt.show()
```

Create a scatterplot that shows the relationship between horsepower and weight.

```python
sns.scatterplot(x='horsepower', y='weight', data=auto_df)
plt.xlabel('Horsepower')
plt.ylabel('Weight')
plt.title('Relationship between Horsepower and Weight')
plt.show()
```
