```python
import pandas as pd
import numpy as np
```

Below, I created a small dataset of a teacher grading her students and the results will print out just the test results.

```python
students = ['Student 1', 'Student 2', 'Student 3']
tests = ['Test 1', 'Test 2', 'Test 3']
data = np.random.randint(70, 101, size=(3, 3))
print(data)
```

Tidying up the Data

```python
df = pd.DataFrame(data, index=students, columns=tests)
print("Test scores for each student:")
print(df)
```

After tidying up everything, I calculated the average of each students test score.

```python
df = pd.DataFrame(data, index=students, columns=tests)
df['Average'] = df.mean(axis=1)
print("Test scores for each student:")
print(df)
```
