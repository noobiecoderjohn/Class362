```python
import pandas as pd
df = pd.read_csv('Movies.csv')
```

I printed the Dataframe from the csv to make sure it was working properly.

```python
print("DataFrame with ratings by user information:")
print(df)
print()
```

Calculated the average rating by each person and each movie.

```python
average_ratings_user = df.set_index('Name').mean(axis=1)
average_ratings_movie = df.mean()

print("Average ratings for each user:")
print(average_ratings_user)
print()

print("Average ratings for each movie:")
print(average_ratings_movie)
print()
```
