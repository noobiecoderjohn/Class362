Tidying up all the data and calculating the total votes of each candidate based off each time zone brackets.

```python
candidates = ['Candidate A', 'Candidate B', 'Candidate C']
timezones = ['Eastern', 'Central', 'Mountain', 'Pacific', 'Alaska', 'Hawaii']
votes_data = np.random.randint(10000, 50000, size=(len(candidates), len(timezones)))

votes_df = pd.DataFrame(votes_data, index=candidates, columns=timezones)
votes_df['Total Votes'] = votes_df.sum(axis=1)

print("Vote data for each candidate:")
print(votes_df)
```
